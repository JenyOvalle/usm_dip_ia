#1. TRABAJANDO CON PANDAS

>1.1 Importe la librería pandas.

>1.2 Cargue el archivo estudiantes y conviertalo en un DataFrame

>1.3 Imprima el contenido del archivo.

>1.4 Aplique estadística descriptiva para:
>> Calcular el promedio total de las notas.
>> Encontrar la nota mayor y mínima.
>> Calcular el promedio de la edad.
>> Mostrar todas las notas mayores o iguales a 55.

>1.5 Cree tablas de contingencia para:
>>Mostrar la frecuencia entre género y modalidad de clase.
>>Mostrar la frecuencia entre ciudad y modalidad de clases.
>>Mostrar la frecuencia entre género, ciudad y modalidad de clases.

>1.6 Flitrar datos
>>Mostrar los datos de estudiantes de género femenino con promedio igual o mayor a 55.

>>Mostrar datos de estudintes con promedio mayor o igual a 55.

>>Mostrar los datos de todos los estudiantes que vienen de Coronel.

>1.7 Convertir DataFrame a un archivo
>>Guarde todos los datos de estudiantes aprobados en un DataFrame llamado aprobados.
>>Convierta el DataFrame al archivo aprobados.csv y guardelo en su equipo.


```python
import pandas as pd

# 1.2 Cargar el archivo estudiantes y convertirlo en un DataFrame
file_path = "/content/estudiantes.xlsx"  # Asegúrate de cambiar esta ruta según la ubicación de tu archivo
estudiantes = pd.read_excel(file_path)

# 1.3 Imprimir el contenido del archivo
print("Contenido del archivo estudiantes:")
print(estudiantes)

# 1.4 Aplicar estadística descriptiva
# Calcular el promedio total de las notas
promedio_notas = estudiantes['PromFinal'].mean()
print(f"Promedio total de las notas: {promedio_notas}")

# Encontrar la nota mayor y mínima
nota_max = estudiantes['PromFinal'].max()
nota_min = estudiantes['PromFinal'].min()
print(f"Nota mayor: {nota_max}, Nota mínima: {nota_min}")

# Calcular el promedio de la edad
promedio_edad = estudiantes['Edad'].mean()
print(f"Promedio de la edad: {promedio_edad}")

# Mostrar todas las notas mayores o iguales a 55
notas_mayores_55 = estudiantes[estudiantes['PromFinal'] >= 55]
print("Notas mayores o iguales a 55:")
print(notas_mayores_55)

# 1.5 Crear tablas de contingencia
# Frecuencia entre género y modalidad de clase
frecuencia_genero_modalidad = pd.crosstab(estudiantes['Genero'], estudiantes['Modalidad'])
print("Frecuencia entre género y modalidad de clase:")
print(frecuencia_genero_modalidad)

# Frecuencia entre ciudad y modalidad de clase
frecuencia_ciudad_modalidad = pd.crosstab(estudiantes['Ciudad'], estudiantes['Modalidad'])
print("Frecuencia entre ciudad y modalidad de clase:")
print(frecuencia_ciudad_modalidad)

# Frecuencia entre género, ciudad y modalidad de clase
frecuencia_genero_ciudad_modalidad = pd.crosstab([estudiantes['Genero'], estudiantes['Ciudad']], estudiantes['Modalidad'])
print("Frecuencia entre género, ciudad y modalidad de clase:")
print(frecuencia_genero_ciudad_modalidad)

# 1.6 Filtrar datos
# Estudiantes de género femenino con promedio igual o mayor a 55
femenino_mayor_55 = estudiantes[(estudiantes['Genero'] == 'F') & (estudiantes['PromFinal'] >= 55)]
print("Estudiantes de género femenino con promedio igual o mayor a 55:")
print(femenino_mayor_55)

# Estudiantes con promedio mayor o igual a 55
promedio_mayor_55 = estudiantes[estudiantes['PromFinal'] >= 55]
print("Estudiantes con promedio mayor o igual a 55:")
print(promedio_mayor_55)

# Estudiantes que vienen de Coronel
estudiantes_coronel = estudiantes[estudiantes['Ciudad'] == 'Coronel']
print("Estudiantes que vienen de Coronel:")
print(estudiantes_coronel)

# 1.7 Convertir DataFrame a un archivo
# Guardar los datos de estudiantes aprobados en un DataFrame llamado aprobados
aprobados = estudiantes[estudiantes['PromFinal'] >= 55]

# Convertir el DataFrame al archivo "aprobados.csv"
aprobados.to_csv("aprobados.csv", index=False)
print("El archivo 'aprobados.csv' ha sido guardado con los datos de estudiantes aprobados.")
```

#2. TRABAJANDO CON MATPLOTLIB
>2.1 Importe la librería.

>2.2 Use el mismo DataFrame de la parte 1.

>2.3 Grafique los datos de género y edad con un gráfico de barras.

>2.4 Grafique los datos de edad y modalidad con un gráfico de dispersión.

>2.5 Grafique los datos de modalidad y Promediofinal con un gráfico de línea.


```python
import pandas as pd
import matplotlib.pyplot as plt

# 2.2 Usar el mismo DataFrame del ejercicio anterior
file_path = "estudiantes.xlsx"  # Asegúrate de cambiar esta ruta según la ubicación de tu archivo
estudiantes = pd.read_excel(file_path)

# 2.3 Graficar los datos de género y edad con un gráfico de barras
genero_edad = estudiantes.groupby('Genero')['Edad'].mean()
genero_edad.plot(kind='bar', title='Promedio de Edad por Género')
plt.xlabel('Género')
plt.ylabel('Edad Promedio')
plt.show()

# 2.4 Graficar los datos de edad y modalidad con un gráfico de dispersión
plt.scatter(estudiantes['Edad'], estudiantes['Modalidad'].astype(str))
plt.title('Relación entre Edad y Modalidad de Clase')
plt.xlabel('Edad')
plt.ylabel('Modalidad')
plt.show()

# 2.5 Graficar los datos de modalidad y Promediofinal con un gráfico de línea
promedio_modalidad = estudiantes.groupby('Modalidad')['PromFinal'].mean()
promedio_modalidad.plot(kind='line', marker='o', title='Promedio de Notas por Modalidad')
plt.xlabel('Modalidad')
plt.ylabel('Promedio de Notas')
plt.show()
```

#3. TRABAJANDO CON SCIKIT-LEARN
>3.1 Importe la librería.

>3.2 Use el mismo DataFrame de la parte 1.

>3.3 Cree un DataFrame con las edades normalizadas para que tengan una media de cero y una desviación estandar de 1.

>3.4 Binarice la columna promedio según un umbral mayor o igual que 55. Notas menores que ese umbral tienen el valor 0, en caso contrario, tienen el valor 1.

>3.5 Codifique la columna género usando one-hot encoding

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, Binarizer, OneHotEncoder

# 3.2 Usar el mismo DataFrame del primer ejercicio
file_path = "estudiantes.xlsx"  # Asegúrate de cambiar esta ruta según la ubicación de tu archivo
estudiantes = pd.read_excel(file_path)

# 3.3 Crear un DataFrame con las edades normalizadas
scaler = StandardScaler()
edades_normalizadas = scaler.fit_transform(estudiantes[['Edad']])
edades_normalizadas_df = pd.DataFrame(edades_normalizadas, columns=['edad_normalizada'])
print("Edades normalizadas:")
print(edades_normalizadas_df)

# 3.4 Binarizar la columna promedio según el umbral >= 55
binarizer = Binarizer(threshold=55)
binario_promedio = binarizer.fit_transform(estudiantes[['PromFinal']])
estudiantes['promedio_binarizado'] = binario_promedio
print("Columna promedio binarizada:")
print(estudiantes[['PromFinal', 'promedio_binarizado']])

# 3.5 Codificar la columna género usando one-hot encoding
encoder = OneHotEncoder(sparse_output=False)
genero_encoded = encoder.fit_transform(estudiantes[['Genero']])
genero_encoded_df = pd.DataFrame(genero_encoded, columns=encoder.get_feature_names_out(['Genero']))
print("Columna género codificada (one-hot encoding):")
print(genero_encoded_df)


# Combinar el DataFrame original con los nuevos datos
estudiantes_preprocesado = pd.concat([estudiantes, edades_normalizadas_df, genero_encoded_df], axis=1)
print("DataFrame preprocesado:")
print(estudiantes_preprocesado)
```
